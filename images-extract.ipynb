## cleaning data--> removing stop words, punctuations etc

```python
import nltk
from nltk.corpus import stopwords
import string
from nltk.stem import WordNetLemmatizer
from nltk.corpus import wordnet
```

```python
stops = set(stopwords.words('english'))
punc = list(string.punctuation)
stops.update(punc)
lemmatizer = WordNetLemmatizer()
```

```python
import re
from nltk import pos_tag
```

```python
function to clean the transcript
```

```python
def get_clean(w):
    clean = []
    w = re.sub(r'http\S+', '', w) 
    w = re.sub(r'@[A-Za-z0-9]+', '', w) 
    w = re.sub(r'#', '', w)
    if w.lower() not in stops:
        pos = pos_tag([w])
        w = ' '.join([lemmatizer.lemmatize(i ) for i in w.split()])
    
    return w
```

```python
df_extracted_data['transcript']=df_extracted_data['transcript'].apply(get_clean)
```
